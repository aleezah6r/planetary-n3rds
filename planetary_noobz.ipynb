```python
import numpy as np

import astropy.units as u
from astropy.time import Time
from astropy.coordinates import SkyCoord
import pytz

from astroplan import Observer, FixedTarget
from astropy.coordinates import get_sun, get_body, get_moon

from astroquery.simbad import Simbad

%matplotlib inline
import matplotlib.pyplot as plt
from astroplan.plots import plot_sky, plot_airmass
```

```python
Helix = FixedTarget.from_name("NGC 7293")
Ring = FixedTarget.from_name("NGC 6720")
Cats = FixedTarget.from_name("NGC 6543")
Ghost = FixedTarget.from_name("NGC 6364")
Eskimo = FixedTarget.from_name("NGC 2392")
targets = np.array([Helix, Ring, Cats, Ghost, Eskimo])
```

```python
apo =  Observer.at_site('apo')
observation_dates = np.array([Time('2017-04-17'),
                              Time('2017-04-20'),
                              Time('2017-04-24'),
                              Time('2017-05-01')])
local_time_zone = pytz.timezone('US/Mountain')

for i in targets:        
    for j in observation_dates:
        if apo.target_is_up(j,i) == True:
            start_time = apo.sun_set_time(j).to_datetime(local_time_zone)
            end_time = apo.sun_rise_time(j).to_datetime(local_time_zone)
            delta_t = end_time - start_time
            observe_time = start_time + delta_t * np.linspace(0.0, 1.0, 20)
            plot_sky(i, apo, observe_time)
            
            airmass_my_target = apo.altaz(observe_time, i).secz
            print('')
            print(i.name,
                  apo.target_is_up(j,i))
            
            print('')
            for idx, mass in enumerate(airmass_my_target):
                if mass <= 1.5 and mass > 0:
                    output_string = "At {0} the airmass is {1}".format(observe_time[idx],mass)
                    print(output_string)
            
    plt.show()

```
